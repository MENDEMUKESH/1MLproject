```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('carprices.csv')
```

```python
df
```

```python
X = df[['Car Model','Mileage','Age(yrs)']]
```

```python
X
```

```python
y= df['Sell Price($)']
y
```

```python
dummies= pd.get_dummies (df['Car Model'],dtype='int')
dummies
```

```python
merged =pd.concat([X,dummies],axis ='columns')
merged
```

```python
final = merged.drop(['Car Model'],axis ='columns')
final
```

```python
final =final.drop(['BMW X5'],axis='columns')
final
```

```python
reg  = linear_model.LinearRegression()
reg.fit(final,y)
```

```python
reg.predict([[69000,6,0,0]])
```

```python
reg.score(final,y)
```
