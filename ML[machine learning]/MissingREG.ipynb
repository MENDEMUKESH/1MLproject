```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('hiring.csv')
```

```python
df
```

```python
df.rename({'test_score(out of 10)':'test_score','interview_score(out of 10)':'interview_score'},axis=1,inplace=True)
```

```python
df
```

```python
df.experience.fillna('zero',inplace=True)
```

```python
import math
mean=math.floor(df.test_score.mean())
mean
```

```python
df.test_score.fillna(mean,inplace=True)
```

```python
df
```

```python
from word2number import w2n
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
df.experience
```

```python
reg=linear_model.LinearRegression()
X=df[['experience','test_score','interview_score']]
y=df['salary($)']
reg.fit(X,y)
```

```python
reg.predict([[1,2,3]])
```
