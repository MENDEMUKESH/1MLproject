```python
train_X = train_X/255
test_X = test_X/255
```

```python
train_X_new = train_X.reshape(len(train_X),28*28)
train_X_new.shape
```

#Building  a neural network from here

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation ='sigmoid'),
    keras.layers.Dense(10,activation ='softmax'),
    keras.layers.Dense(10,activation ='softmax')


])
model.compile(
    optimizer = 'adam',
    loss ='SparseCategoricalCrossentropy',
    metrics =['accuracy']
)
model.fit(train_X_new,train_y,epochs=10)
```

```python
model.evaluate (test_X_new,test_y)
```
