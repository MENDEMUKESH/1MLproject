```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset

df = pd.read_csv("license_plate_dataset.csv")
df


```

```python
# Histogram: Distribution of Region Codes
plt.figure(figsize=(10, 6))
df['Region_Code'].plot(kind='hist', bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Region Codes')
plt.xlabel('Region Code')
plt.ylabel('Frequency')
plt.show()
```

```python
number= cv2.imread('img2.jpeg')
plt.show()
plt.imshow(number)
```

```python

# Scatter Plot: Relationship between Region Code and Vehicle Type
plt.figure(figsize=(10, 6))
plt.scatter(df['Region_Code'], df['Vehicle_Type'], color='red', alpha=0.5)
plt.title('Scatter Plot of Region Code vs. Vehicle Type')
plt.xlabel('Region Code')
plt.ylabel('Vehicle Type')
plt.show()
```

```python
# Pie Chart: Proportion of Different Vehicle Types
plt.figure(figsize=(8, 8))
df['Vehicle_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Proportion of Different Vehicle Types')
plt.ylabel('')  # Hides the y-label
plt.show()
```

```python
# Bar Graph: Distribution of Vehicle Types
plt.figure(figsize=(10, 6))
df['Vehicle_Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.show()
```

```python
# Line Graph: Cumulative count of Plate Numbers by Vehicle Type
plt.figure(figsize=(10, 6))
df.groupby('Vehicle_Type').size().cumsum().plot(kind='line', marker='o', color='purple')
plt.title('Cumulative Count of Plate Numbers by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Cumulative Count')
plt.grid(True)
plt.show()
```
